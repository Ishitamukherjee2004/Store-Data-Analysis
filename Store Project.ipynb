```python
dataset["Ship Mode"].value_counts()
```

```python
plt.pie(x = dataset["Ship Mode"].value_counts().values, labels = dataset["Ship Mode"].value_counts().index, autopct = "%1.2f%%", startangle = 80)
plt.legend(loc=2)
plt.show()
```

```python
sns.countplot(x="Ship Mode", data = dataset, hue = "Category")
plt.show()
```

```python
sns.countplot(x="Category", data = dataset)
plt.show()
```

```python
plt.figure(figsize=(14, 4))
sns.countplot(x="Category", data = dataset, hue = "Sub-Category")
plt.show()
```

```python
plt.figure(figsize=(14, 4))
sns.countplot(x="Category", data = dataset[dataset["Category"]=="Office Supplies"], hue = "Sub-Category")
plt.show()
```
